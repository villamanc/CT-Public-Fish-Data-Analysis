```python
import os
import pandas as pd
```

## Step 1 Combine All Town Level Data Into One DF

```python
#Get the current dir
current_directory = os.getcwd()

#Initialize an empty list to store dfs
dfs = []
```

```python
#Loop through all files in the directory
for file in os.listdir(current_directory):
    #Check if the file is a CSV file
    if file.endswith(".csv"):
        #Read the CSV file into a df
        df = pd.read_csv(os.path.join(current_directory, file))
        #Extract the town name from the filename
        town_name = os.path.splitext(file)[0]
        #Add a new column with the town name
        df['Town'] = town_name
        #Reorder columns to place 'Town' at the 5th position
        cols = list(df.columns)
        cols.insert(4, cols.pop(cols.index('Town')))
        df = df[cols]
        #Append the df to the list
        dfs.append(df)
```

```python
#Combine all dfs in the list into a single df
df = pd.concat(dfs, ignore_index=True)
```

## Step 2 Remove NA Values

```python
df
```

```python
df.isna().sum()
```

```python
#drop na values
df = df.dropna()
```

## Step 3 Add Geographic Data

```python
#Insert a column with the state
df.insert(4, 'State', 'Connecticut')
df
```

## Step 4 Remove Unwanted/Irrelevant Fish Species

```python
#Select columns all fish columns
total_fish = df.iloc[:, 6:]

#Calculate the sum of each column
total_fish = total_fish.sum(axis=0)

#Create a df with the totals for each column and a single row
total_fish = pd.DataFrame([total_fish])

#Melt the df
total_fish = total_fish.melt(var_name='Type of Fish', value_name='Count')

#Sort the df from least to greatest
total_fish = total_fish.sort_values(by='Count',ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Create the horizontal bar plot
plt.figure(figsize=(10, 30))  # Adjust the figure size as needed
sns.barplot(x='Count', y='Type of Fish', data=total_fish, color='skyblue')

#Set the title and labels
plt.title('Total Fish Found By Count')
plt.xlabel('Total Number Found In Data Set')
plt.ylabel('Type of Fish')

#Show the chart of fish counts by species
plt.show()
```

```python
#Save 36 of the 90 unique species to anazlyze
filtered_total_fish = total_fish.head(36)

filtered_total_fish
```

```python
#Drop the count column from the fish df
filtered_total_fish = filtered_total_fish.drop(columns=['Count'])

#Make a list of all the unique fish in the fish df
fish_columns = [item for sublist in filtered_total_fish.values.tolist() for item in sublist]

#Make a list of all the columns that come before the fish columns
all_columns = df.iloc[:, 0:6].columns.tolist()

#Make a list of all the columns by adding all the columns in the fish_columns list to the all_columns varaible
for col in fish_columns:
    all_columns.append(col)


#Make a df of only the columns in the all_columns list
total_fish = df[all_columns]
```

## Step 5 Melt The Data Frame For Excel

```python
#Melt the df so all fish count columns go to a "Fish Type" column, and all counts to a "Count" column
melted_df = pd.melt(df, id_vars=total_fish.iloc[:,0:6].columns.tolist(), var_name='Fish Type', value_name='Count')
melted_df
```

```python
#Filter out any columns where the count is 0
melted_df = melted_df[melted_df['Count'] != 0]
melted_df
```

```python
#Turn the df to an Excel file
melted_df.to_excel("Clean Data.xlsx",index=False)
```
